## Question 5

What is MapReduce? How is it related to Hadoop?

Write your answer below this line: 
_______________________________________________________________________________________________________________________________The idea of having each server run some query or function at the same time, and then combining the results is Map Reduce.  When we ask each server to run the function or query at the same time, this is the "Map" step. When we combine the results from each server into a single aggregate, this is the "Reduce" step.  Hadoop is a program used for RDD's or Big Data.
